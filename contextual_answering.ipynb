## Text extraction

```python
%pip install pdfminer.six
```

```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer

def extract_text_from_pdf_page(file_path):
    text = ''
    for page_layout in extract_pages(file_path):
        if page_layout.pageid > 2:
            for element in page_layout:
                if isinstance(element, LTTextContainer):
                    text += element.get_text()
    return text

pdf_file = "monitoring.pdf"
text = extract_text_from_pdf_page(pdf_file)
print(text)
```

## Text cleaning

```python
# Strings to remove
strings_to_remove = ['© 2020 JERA Co., Inc. All Rights Reserved.', 'Evaluation only.', '取扱厳秘', 'Created with Aspose.Slides for .NET Standard 2.0 22.7.', 'Copyright 2004-2022Aspose Pty Ltd.','\n\n\n']

# Remove all occurrences of the strings
for string in strings_to_remove:
    text = text.replace(string, '')

# Characters to replace
characters_to_replace = ['§', 'q','ü']

# Replace characters with hyphen
for char in characters_to_replace:
    text = text.replace(char, '-')

# Print the extracted text
print(text)
```

## QA with pipeline

```python
from transformers import pipeline

qa_model = pipeline("question-answering")
```

```python
question = "What are the SAP Business Process Operations?"
qa_model(question = question, context = text)
```

## QA with distilbert

```python
import torch
from transformers import DistilBertForQuestionAnswering, DistilBertTokenizer
```

```python
# Question to ask
question = "What are the SAP Business Process Operations?"

# Load the pre-trained question-answering model and tokenizer
model = DistilBertForQuestionAnswering.from_pretrained('distilbert-base-uncased-distilled-squad')
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased-distilled-squad')

```

```python
# Tokenize the input text and question
inputs = tokenizer.encode_plus(question, text, add_special_tokens=True, truncation=True, max_length=512, padding='max_length', return_tensors='pt')

# Forward pass through the model
outputs = model(**inputs)

# Get the start and end logits from the model's output
start_logits = outputs.start_logits
end_logits = outputs.end_logits

# Get the most probable answer using the start and end logits
start_index = torch.argmax(start_logits, dim=1).item()
end_index = torch.argmax(end_logits, dim=1).item()
answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs['input_ids'][0, start_index:end_index+1].tolist()))

# Print the answer
print("Answer:", answer)
```

## QA with paragraphs

```python
import re

def delete_numeric_lines(text):
    lines = text.split('\n')
    filtered_lines = []
    for line in lines:
        # Use regular expression to check if line contains only numbers
        if not re.match(r'^\d+$', line.strip()):
            filtered_lines.append(line)
    filtered_text = '\n'.join(filtered_lines)
    return filtered_text
```

```python
from textblob import TextBlob

def split_into_paragraphs(text):
    blob = TextBlob(text)
    paragraphs = blob.raw.split('Page')  # Split based on double newlines
    return paragraphs
paragraphs = split_into_paragraphs(text)

for paragraph in paragraphs:
    paragraph = delete_numeric_lines(paragraph)
    print(paragraph)
    print('-' * 20)  # Separator between paragraphs
```

## <u>Sentence generation</u>

## Transformers

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Load the pre-trained "distilgpt2" model and tokenizer
model = GPT2LMHeadModel.from_pretrained('distilgpt2')
tokenizer = GPT2Tokenizer.from_pretrained('distilgpt2')
```

```python

# Set the generated sentence length
sentence_length = 20

# Generate a sentence based on the answer
input_ids = tokenizer.encode(ans["answer"], return_tensors='pt')
attention_mask = input_ids.clone().fill_(1)  # Set attention mask to 1 for all input tokens

# Generate a sentence based on the answer
generated_sentence = model.generate(
    input_ids=input_ids,
    attention_mask=attention_mask,
    max_length=sentence_length,
    num_return_sequences=1,
    no_repeat_ngram_size=2,
    early_stopping=True
)

# Decode the generated sentence
decoded_sentence = tokenizer.decode(generated_sentence[0], skip_special_tokens=True)

# Print the generated sentence
print("Generated Sentence:", decoded_sentence)
```
